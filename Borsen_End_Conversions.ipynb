##### Sessions

```python
# See the first line
with open('data/borsen_sessions_2017101-20171201.csv', encoding="utf8") as fp:
    print(fp.readline())
```

```python
# Sessions
session = pd.read_csv('data/borsen_sessions_2017101-20171201.csv', delimiter='\t', 
                      parse_dates=['SessionStart', 'ServedTime', 'ClosedTime', 'ConversionTime'], 
                      index_col=['SessionStart'], dtype={'Rule': 'category', 'ConversionType':'category'})
# Time interval
begin, end = session.sort_index().index[[0,-1]].date.astype(str)
print('From {:} to {:}'.format(begin, end))
```

```python
session
```

```python
session.shape
```

```python
# Select only served
session=session[~session['ServedTime'].isnull()]
```

```python
# Select only for active rules
session = session[session['Active']]
```

```python
# Some conversions
session['Active'] = session['Active'].astype('bool')
session['CPID'] = session['CPID'].astype('int')
```
